```python
x = df.iloc[:,:].values
```

```python
x.shape
```

```python
# Elbow Method
from sklearn.cluster import KMeans
wcss = [] #within cluster sum of square
for i in range(1, 18):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=17, max_iter=375, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

Importing KMeans: from sklearn.cluster import KMeans
Initializing WCSS List: wcss = [] - WCSS stands for "Within-Cluster Sum of Squares" which measures the variance within each cluster.
Loop for Different Clusters: for i in range(1, 18): - Looping through a range of cluster numbers (from 1 to 17 in this case).
KMeans Configuration: kmeans = KMeans(n_clusters=i, init='k-means++', n_init=17, max_iter=375, random_state=0) - Setting up the KMeans algorithm:
n_clusters=i: Number of clusters in the current iteration.
init='k-means++': Method for initialization.
n_init=17: Number of times the algorithm will be run with different centroid seeds.
max_iter=375: Maximum number of iterations for a single run.
random_state=0: Ensures reproducibility of results.
Fitting the Model: kmeans.fit(x) - Training the KMeans algorithm on the data x.
Appending WCSS: wcss.append(kmeans.inertia_) - Appending the WCSS (inertia) value of the current model to the wcss list.

```python
plt.plot(range(1,18), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# fitting kmeans to dataset
kmeans = KMeans(n_clusters=2, init='k-means++', n_init=17, max_iter=375, random_state=18)
Y_kmeans = kmeans.fit_predict(x)
```

KMeans clustering model  applied to a dataset and predicts the cluster assignments for each data point.
n_clusters=2: Specifies the number of clusters to form (in this case, 2).
init='k-means++': Method for initializing the centroids to speed up convergence.
n_init=17: Number of times the KMeans algorithm will be run with different centroid seeds. The final results will be the best output of these runs in terms of inertia.
max_iter=375: Maximum number of iterations the algorithm will run for a single run.
random_state=18: Seed used by the random number generator to ensure reproducibility.
kmeans.fit_predict(x): Fits the KMeans model to the data x and returns the cluster assignments for each data point in x.
Y_kmeans: Array of cluster labels assigned to each data point.

```python
from sklearn import metrics
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
```

**silhouette_score**

Description: Measures how similar an object is to its own cluster (cohesion) compared to other clusters (separation).
Range: -1 to 1, where a higher value indicates better clustering.
Interpretation: A value close to 1 means the clusters are well-separated, while a value close to -1 indicates overlapping clusters.

**calinski_harabasz_score**

Description: Also known as the Variance Ratio Criterion, it evaluates the ratio of the sum of between-cluster dispersion to within-cluster dispersion.
Higher is better: A higher score indicates more distinct and well-separated clusters.



**davies_bouldin_score**

Description: Measures the average similarity ratio of each cluster with the cluster that is most similar to it, based on the ratio of within-cluster distances to between-cluster distances.
Lower is better: A lower score indicates better clustering with more distinct clusters.

```python
print(silhouette_score(x,kmeans.labels_))
print(calinski_harabasz_score(x,kmeans.labels_))
print(davies_bouldin_score(x,kmeans.labels_))
```
